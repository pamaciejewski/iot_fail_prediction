```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

# Load the data
data = pd.read_csv('../data/iot_sensor_dataset.csv')
data.head()
```

```python
# Display the shape of the dataset and check for missing values
print("Data shape:", data.shape)
print("\nMissing values in each column:\n", data.isnull().sum())

data.describe()
```

```python
# Histograms
data.hist(figsize=(10, 10), bins=20)
plt.tight_layout()
plt.show()

# Boxplots to check for outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.xticks(rotation=90) 
plt.title('Boxplot for numerical variables')
plt.show()
```

```python
correlation_matrix = data.corr() 

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
for column in data.select_dtypes(include=['float', 'int']).columns:
    sns.histplot(data[column], kde=True)
    plt.title(f'Histogram and KDE for {column}')
    plt.show()
```

```python
for column in data.select_dtypes(include=['float', 'int']).columns:
    sns.ecdfplot(data[column])
    plt.title(f'Cumulative Distribution Function for {column}')
    plt.show()
```
